```python
def func(M,M1,phi1,a):
    y=[]
    for i in M:
        y1=0.4*log(10)*phi1*(10**(-0.4*(i-M1)*(a+1)))*exp(-10**(-0.4*(i-M1)))
        y.append(y1)
    
    return(y)  
```

```python
def likelihood(M,phi,M1,phi1,a,s): #phi1,M1 y a son los parametros de la funcion de schester,y son los datos dados por la funcion y phi son los blanton
    z=func(M,M1,phi1,a)
    x=((phi - z)/s)**2
    x2=np.sum(x)
    l=np.exp(-x2/2)
    return l
```

```python
print(phi)
likelihood(M,phi,-20.83,0.0146,-1.20,s)
```

```python
philim = [0.00146,0.146]
alim = [-2.20,-0.20]
Mlim = [-21.83,-19.83]

def prior(philim,alim,Mlim,a,M1,phi1):
    if (a<=alim[1]) & (a>=alim[0]) & (M1<=Mlim[1]) &( M1>=Mlim[0]) & (phi1<=philim[1]) & (phi1>=philim[0]):
        return 1
    else:
        return 0
```

```python
prior(philim,alim,Mlim,-2,-20,0.01)
```

```python
def posterior(philim,alim,Mlim,a,M1,phi1,M,phi):
    p=prior(philim,alim,Mlim,a,M1,phi1)*likelihood(M,phi,M1,phi1,a,s)
    return p
```

```python
N=10000

l=0
while l==0:
    phi1=np.random.uniform(0.00146,0.146)
    M1=np.random.uniform(-21.83,-19.83)
    a=np.random.uniform(-2.20,-0.20)
        
    l=likelihood(M,phi,M1,phi1,a,s)
    phi10,M10,a0 = phi1,M1,a
#print(phi10,M10,a0)

l0=likelihood(M,phi,M10,phi10,a0,s)
p0=posterior(philim,alim,Mlim,a0,M10,phi10,M,phi)
#print(l0)
a= [a0]
phi_lista=[phi10]
m=[M10]
l_list=[l0]
p_list=[p0]
for i in range(N-1):
    #print(i)
    a_new=np.random.normal(a[i],0.001)
    m_new=np.random.normal(m[i],0.01)
    phi_new=np.random.normal(phi_lista[i],0.0001)
    l_new=likelihood(M,phi,m_new,phi_new,a_new,s)
    p_new=posterior(philim,alim,Mlim,a_new,m_new,phi_new,M,phi)
    #print(l_new)
    l_list.append(l_new)
    p_list.append(p_new)
    
    c=min(1,p_list[i+1]/p_list[i])
    u=np.random.uniform(0,1)
    if u<=c:
        a0=a_new
        M10=m_new
        phi10=phi_new
        
        
    a.append(a0)
    m.append(M10)
    phi_lista.append(phi10)

            
#print(a)
```
